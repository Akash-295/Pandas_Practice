### Pandas Lesson 2: Selecting, Filtering, and Accessing Data

```python
import pandas as pd 

data = {
    "Name": ['Shinra','Tanjiro','Arthor','Rey','Kunon'],
    "Age": [20,21,19,20,19],
    "Marks": [89, 78, 67, 97, 90],
    "City": ['Delhi','Dehradun','Delhi','Gurgaon','Roorkee']  
}

df = pd.DataFrame(data)
print(df)
```

```python
#print only name column
print(df['Name'])
```

```python
#print name and marks column
print(df[["Name","Marks"]])
```

```python
#print row index 3 using loc
print(df.loc[3])
```

```python
#prict students marks > 80 
print(df[df["Marks"]>80])
```

```python
# Print students with Age > 19
# AND Marks > 85
print(
    df[
        (df["Age"]>19)
         &
        (df["Marks"]>85)
    ])
```

```python
# Print students from Delhi
print(df[
    df["City"] == "Delhi"
])
```

```python
#sort by Marks descending 
print(df.sort_values("Marks", ascending = False))
```

```python
# Count how many students belong to each city
print(df["City"].value_counts())
```

```python
# Highest marks
# Lowest marks
# Average marks
# Total marks

print(df["Marks"].max())
print(df["Marks"].min())
print(df["Marks"].mean())
print(df["Marks"].sum())
```

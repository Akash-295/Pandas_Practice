## Working with Missing DATA in Pandas 

```python
Data = pd.read_csv('abc1.csv')
print(Data)
```

```python
print(Data.isnull().sum())
```

```python
print(Data.dropna())  #it drop complete record not only nan 
```

```python
import numpy as np
print(Data.replace(np.nan, "Hi"))
```

```python
Data["maths"] = Data["maths"].replace(np.nan, 33.6) #fill specif column 
print(Data)
```

```python
print(Data['maths'].mean())
```

```python
print(Data.fillna(65))
```

##Column Transformation 


```python
import pandas as pd
```

```python
df = pd.read_csv('employees.csv')
print(df)

df.loc[(df["Bonus %"] == 0), "GetBonus"] = "no Bonus"
df.loc[(df["Bonus %"] > 0, "GetBonus")] = "Bonus"

print(df.head(10))
```

```python
data = pd.read_csv("abc1.csv")
print(data.fillna(65))

data["total_marks"] = data["language"]+data["science"]+data["maths"]+data["history"]
print(data)
```

```python
data1 = pd.read_csv("employees.csv")
print(data1.head(10))

data1["Bonus"] = (data1['Salary']/100)*20
print(data1.head(10))
```

```python
data2 = {"Months": ["january", "february", "march", "april"]}

a = pd.DataFrame(data2)
print(a)

def extract(value):
    return value[0:3]

a['Short_months'] = a["Months"].map(extract)
print(a)
```

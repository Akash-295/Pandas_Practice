#Pandas Group by 

```python
import pandas as pd

data = pd.read_csv("employees.csv")
print(data)
```

```python
gp = data.groupby("Team").agg({'Gender': 'count'})
print(gp)
```

#merge, join and concatenate in pandas 

```python
data1 = {'Emp_id': ['e01','e02','e03','e04','e05'],
         "Names": ['Ram', 'Rajesh', 'Manpreet','kulvinder','karpa'],
         "Age": [23,34,32,45,29]}

data2 = { 'Emp_id': ['e01','e02','e03','e04','e05'],
         "Salary": [23000, 43000, 26000, 34000, 38000]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print(df2)
```

```python
#merge or concate 

print(pd.merge(df1,df2,on="Emp_id"))
```

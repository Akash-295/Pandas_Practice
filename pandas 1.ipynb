### Pandas Lesson 1: Introduction to Pandas (Deep Dive)

```python
import pandas as pd 

S = pd.Series([100, 200, 300, 400, 500], name = 'Numbers')
print("Series:", S)

print("Shape of Series: ",S.shape)
print("Values of Series: ",S.values)
print("indexes of Series: ",S.index)
```

##Making a Dataframe

```python
data = {
    "Laptop": [50000],
    "Phone": [25000],
    "Tablet": [30000]
}

new = pd.DataFrame(data)
print(new)
print("shape", new.shape)
print("Dtype: ", new.dtypes)
print("columns: ", new.columns)
print(new.info())
```

##Series of marks

```python
S1 = pd.Series([68,86,78,92,89], name = "Marks")
print(S1)
print("Mean:",S1.mean())
print("Maximux: ",S1.max())
print("Minimum: ",S1.min())
```

##Data frame operations

```python
data = {
    "Name":['Akash', 'Anshika', 'Aman', 'Aniket'],
    "Age":[19, 19, 20, 21],
    "city":['Delhi','Dehradun','Roorkee','Delhi']
}

df = pd.DataFrame(data)
print("Data Frame: \n",df)
print("head values: \n",df.head(2))
print("Tail Values: \n", df.tail(2))
print("Shape: \n", df.shape)
```
